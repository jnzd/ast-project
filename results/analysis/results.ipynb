```python
import pandas as pd
```

```python
results_all = pd.read_csv('../findings/i7-1165G7/gcc-10-vs-gcc-12/all-diffs/mutation_attempts.csv')
```

```python
results_all.head()
```

```python
results_diffs = results_all[results_all['asm_diff'] > 0]
```

```python
results_per_seed = results_diffs.groupby('filename').count()
results_per_seed
```

```python
max_results_per_seed = results_diffs.groupby('filename').max('asm_diff')
max_results_per_seed
```

```python
max_results_overview = max_results_per_seed[['mutation-id', 'asm_diff']]
max_results_overview
```

```python
max_results_greater_one = max_results_overview[max_results_overview['asm_diff'] > 1]
max_results_greater_one
```

```python
max_results_greater_one.to_csv('../findings/i7-1165G7/gcc-10-vs-gcc-12/max-diffs-unique-seed/info.csv')
```
